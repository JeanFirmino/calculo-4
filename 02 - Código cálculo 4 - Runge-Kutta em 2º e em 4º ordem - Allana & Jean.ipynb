```
# Método de Runge-Kutta de ordem 2 #
import time
import numpy as np
from prettytable import PrettyTable
# Para importar as bibliotecas necessárias.
tabela = PrettyTable()
# Criando uma tabela.
tabela.title = "Método de Runge-Kutta de ordem 2 - Feito por Allana & Jean"
tabela.field_names = ["i","x0","y0","yn","Tempo de iteração"]
# Montando o cabeçalho da tabela.
xk2 = []
yk2 = []
# xne e yne será usado para o armazenamento e plotagem dos dados.
def fx(x,y):
    return 3*x*(np.exp(2*x)) - 3*y
# Declarando a função.
def rungekutta2(x0,y0,xn,h):
  n = round((xn - x0) / h)
  # Definindo o número de iterações.
  for i in range(n+1):
    start = time.time()
    # Iniciando o contador.
    k1 = fx(x0, y0)
    k2 = fx(x0+h, y0+h*k1)
    yn = y0 + (h/2)*(k1 + k2)
    # Aplicando o método de Runge-Kutta 2ª ordem segundo Pedrosa.
    stop = time.time()
    # Finalizando o contador.
    tabela.add_row([i+1, x0, y0, yn, stop-start])
    # Adicionando os valores encontrados na tabela.
    xk2.append(x0)
    yk2.append(y0)
    # Adicionando valores para montar o gráfico.
    y0 = yn
    # Substituindo y atual para o y calculado.
    x0 += h
    # Adicionando o passo. 
x0 = 0.50
y0 = 0.00
# x e y iniciales.
xn = 3.00
# Calcular até o ponto xn.
h = 0.50
# h é o passo.
rungekutta2(x0,y0,xn,h)
# Chamada do método de Runge-Kutta de segunda ordem.
print(tabela)
# Apresentar tabela com os valores encontrados.
```

```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# importando bibliotecas necessárias.
def f(y,x):
    dydx = 3*x*(np.exp(2*x)) - 3*y
    return dydx
# A equação diferencial utilizada(dy/dx).
y0 = 0.00
# Condição inicial.
x = np.linspace(0.50,3.00)
# Intervalo para x.
y = odeint(f,y0,x)
# Para resolver utilizando bibliotecas profissionais da linguagem. 
plt.figure(figsize=(8, 4))
plt.plot( x, y,'b-', linewidth=3, color='green', label = 'Scipy(biblioteca profissional)')
plt.plot( xk4, yk4,'o-', color='red', label = 'Runge-Kutta 4ª ordem')
plt.plot( xk2, yk2,'o-', color='orange', label = 'Runge-Kutta 2ªordem')
plt.style.use('classic')
plt.title('Soluções obtidas com o método de Runge-Kutta - h = 0.5')
plt.xlabel('X (Intervalo)')
plt.ylabel('Y (Valores obtidos)')
plt.ylim()
plt.xlim(0.50,3.00)
plt.legend(loc='upper left')
plt.grid('True')
plt.show()
# Configurações usadas para criar o gráfico.
```
