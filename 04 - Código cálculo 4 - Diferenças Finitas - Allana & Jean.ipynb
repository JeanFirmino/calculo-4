```
import numpy as np
from sympy import *
from math import*
from prettytable import PrettyTable

# Para importar as bibliotecas necessárias.
tabela = PrettyTable()
# Criando uma tabela.
tabela.title = "Método de Diferenças Divididas Progressiva - Feito por Allana & Jean"
tabela.field_names = ["X","F(x)","F1(x)","F2(x)","Erro 1","Erro 2"]

#x = Symbol('x')
#y = cos( 4*x**3 + x**2 )
#y1 = y.diff( x, n = 1)
#y2 = y1.diff( x, n = 1)

def y1(x):
    return -(12*x**2 + 2*x)*sin(4*x**3 + x**2)
    
def y2(x):
    return -(24*x + 2)*sin(4*x**3 + x**2) - (12*x**2 + 2*x)**2*cos(4*x**3 + x**2)
```

```
# Diferença dividida finita progressiva

def fx(x):
    return np.cos( 4*x**3 + x**2 )

def f11x(x,h):
    return ( fx(x+h) - fx(x) ) / h 

def f21x(x,h):
    return ( fx(x+2*h) - 2*fx(x+h) + fx(x) ) / (h**2)

xi = np.array([0.45,0.65,0.85])
ype1 = []
ype2 = []

h = 0.20


# Para o as primeiras funções de diferença dividida finita progressiva

for i in range(len(xi)):
  
    erro1 = abs(y1(xi[i]) - f11x(xi[i],h))/abs(y1(xi[i]))
  
    erro2 = abs(y2(xi[i]) - f21x(xi[i],h))/abs(y2(xi[i]))
    ype1.append(erro1)
    ype2.append(erro2)
  
    tabela.add_row([xi[i],fx(xi[i]),f11x(xi[i],h),f21x(xi[i],h),erro1,erro2])

print(tabela)
```

```
# Diferença dividida finita regressiva

def fx(x):

    return np.cos( 4*x**3 + x**2 )

def f11x(x,h):

    return ( fx(x) - fx(x-h) ) / h 

def f21x(x,h):

    return ( fx(x-2*h) - 2*fx(x-h) + fx(x) ) / (h**2)

xi = np.array([0.45,0.65,0.85])
h = 0.20

tab = PrettyTable()

# Criando uma tabela.

tab.title = "Método de Diferenças Divididas Regressiva - Feito por Allana & Jean"

tab.field_names = ["X","F(x)","F1(x)","F2(x)","Erro 1","Erro 2"]
yre1 = []
yre2 = []
# para o as primeiras funções de diferença dividida finita regressiva

for i in range(len(xi)):

    erro1 = abs(y1(xi[i]) - f11x(xi[i],h))/abs(y1(xi[i]))

    erro2 = abs(y2(xi[i]) - f21x(xi[i],h))/abs(y2(xi[i]))
    yre1.append(erro1)
    yre2.append(erro2)
    tab.add_row([xi[i],fx(xi[i]),f11x(xi[i],h),f21x(xi[i],h),erro1,erro2])

print(tab)
```

```
# Diferença dividida finita centrada

def fx(x):

    return np.cos( 4*x**3 + x**2 )

def f11x(x,h):

    return ( fx(x+h) - fx(x-h) ) / (2*h)

def f21x(x,h):

    return ( fx(x+h) - 2*fx(x) + fx(x-h) ) / (h**2)

xi = np.array([0.45,0.65,0.85])
h = 0.20

tabe = PrettyTable()
# Criando uma tabela.
tabe.title = "Método de Diferenças Divididas Centrada - Feito por Allana & Jean"
tabe.field_names = ["X","F(x)","F1(x)","F2(x)","Erro 1","Erro 2"]
# para o as primeiras funções de diferença dividida finita centrada
yce2 = []
yce1 =[]
for i in range(len(xi)):

    erro1 = abs(y1(xi[i]) - f11x(xi[i],h))/abs(y1(xi[i]))

    erro2 = abs(y2(xi[i]) - f21x(xi[i],h))/abs(y2(xi[i]))
    yce2.append(erro2)
    yce1.append(erro1)
    tabe.add_row([xi[i],fx(xi[i]),f11x(xi[i],h),f21x(xi[i],h),erro1,erro2])

print(tabe)
```

```
import numpy as np
import matplotlib.pyplot as plt
# importando bibliotecas necessárias.
plt.figure(figsize=(6, 4))
plt.plot( xi, yre1,'o-', linewidth=4, color='green', label = 'Regressiva')
plt.plot( xi, ype1,'o-', linewidth=4, color='red', label = 'Progressiva')
plt.plot( xi, yce1,'o-', linewidth=4, color='blue', label = 'Centrada')
plt.style.use('classic')
plt.title('Erro 1 - Exemplo 1')
plt.xlabel('X (Intervalo)')
plt.ylabel('Y (Valores obtidos)')
plt.ylim()
plt.xlim()
plt.legend(loc='upper left')
plt.grid('True')
plt.show()
# Configurações usadas para criar o gráfico.
```
