```python
import time
import numpy as np
from prettytable import PrettyTable
# Para importar as bibliotecas necessárias.
tabela = PrettyTable()
# Criando uma tabela.
tabela.title = "Método de Euler - Feito por Allana & Jean"
tabela.field_names = ["i", "x0", "y0", "f(x0,y0)", "yn=y0+h*f(x0,y0)", "Tempo de iteração"]
# Montando o cabeçalho da tabela.
xne = []
yne = []
# xne e yne será usado para o armazenamento e plotagem dos dados.
def f(x,y):
    return x*(np.exp( -x**2 -y**2 ))
# Declarando a função.
def euler(x0,y0,xn,h):
    n = round((xn - x0)/h)
    # Definindo o número de iterações.
    for i in range(n+1):
        start = time.time()
        # Iniciando o contador.
        fx = f(x0, y0)
        yn = y0 + ( h*fx )
        # Aplicando o método de Euler.
        stop = time.time()
        # Finalizando o contador.
        tabela.add_row([i+1, x0, y0, fx, yn, stop-start])
        # Adicionando os valores encontrados na tabela.
        xne.append(x0)
        yne.append(y0)
        # Adicionando valores para montar o gráfico.
        y0 = yn
        # Substituindo y atual para o y calculado.
        x0 += h
        # Adicionando o passo.
x0 = 0.50
y0 = 0.0
# x e y iniciales.
xn = 5.0
# Calcular até o ponto xn.
h = 0.50
# h é o passo.
euler(x0,y0,xn,h)
# Chamada do método de Euler.
print(tabela)
# Apresentar tabela com os valores encontrados.
```

```python
import time
import numpy as np
from prettytable import PrettyTable
# Para importar as bibliotecas necessárias.
tabela = PrettyTable()
# Criando uma tabela.
tabela.title = "Método de Euler modificado - Feito por Allana & Jean"
tabela.field_names = ["i", "x0", "y0", "f(x0,y0)", "y0n=y0+h*fx", "yn=y0+((fx+f(xf,y0n))/2)*h", "Tempo de iteração"]
# Montando o cabeçalho da tabela.
xnem = []
ynem = []
# xne e yne será usado para o armazenamento e plotagem dos dados.
def f(x,y):
    return x*(np.exp( -x**2 -y**2 ))
# Declarando a função.
def eulermodificado(x0,y0,xn,h):
    n = round((xn - x0) / h)
    # Definindo o número de iterações.
    for i in range(n+1):
        start = time.time()
        # Iniciando o contador.
        fx = f(x0, y0)
        y0n = y0 + ( h * fx )
        # Preditor.
        xf = x0 + h
        yn = y0 + ( fx + f(xf, y0n) )*(h/2) # Corretor.
        # Aplicação do método de Euler modificado (preditor-corretor).
        stop = time.time()
        # Finalizando o contador.
        tabela.add_row([i+1, x0, y0, fx, y0n, yn, stop-start])
        # Adicionando os valores encontrados na tabela.
        xnem.append(x0)
        ynem.append(y0)
        # Adicionando valores para montar o gráfico.
        y0 = yn
        # Substituindo y atual para o y calculado.
        x0 += h
        # Adicionando o passo.
x0 = 0.50
y0 = 0.0
# x e y iniciales.
xn = 5.0
# Calcular até o ponto xn.
h = 0.50
# h é o passo.
eulermodificado(x0,y0,xn,h)
# Chamada do método de Euler modificado.
print(tabela)
# Apresentar tabela com os valores encontrados.
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# importando bibliotecas necessárias.
def f(y,x):
    dydx = x*(np.exp( -x**2 -y**2 ))
    return dydx
# A equação diferencial utilizada(dy/dx).
y0 = 0.0
# Condição inicial.
x = np.linspace(0.5,5.0)
# Intervalo para x.
y = odeint(f,y0,x)
# Para resolver utilizando bibliotecas profissionais da linguagem. 
plt.figure(figsize=(10, 4))
plt.plot( x, y,'b-', linewidth=3, color='green', label = 'Scipy(biblioteca profissional)')
plt.plot( xne, yne,'o-', color='red', label = 'Euler')
plt.plot( xnem, ynem,'o-', color='orange', label = 'Euler modificado')
plt.style.use('classic')
plt.title('Soluções obtidas para a EDO estudada.')
plt.xlabel('X (intervalo)')
plt.ylabel('Y (Valores obtidos)')
plt.ylim()
plt.xlim(0.5,5.0)
plt.legend(loc = 'lower right')
plt.grid('True')
plt.show()
# Configurações usadas para criar o gráfico.
```
